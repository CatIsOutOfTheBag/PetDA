# Группировка категориальный переменных

```python
# объединим некоторые категории 'education_level'
df.groupby('education_level')['target'].agg(['count', 'mean'])
```

Способ 1. Написать функцию и применить методом apply к датафрейму

```python
def education_group(x):
  if x in ['High School', 'Primary School']:
    return 'School'
  if x in ['Masters', 'Phd']:
    return 'Masters_and_phd'
  return x
```

```python
df['new_education_level'] = df['education_level'].apply(education_group)
```

```python
df.groupby('new_education_level', dropna=False)['target'].agg(['count', 'mean'])
```

Способ 2. С помощью трюков

```python
df['education_level'].replace('Primary school', 'School')
```

Способ 3. Через словарь

```python
d = {'Primary school': 'School',
     'High School': 'School',
     'Masters': 'Masters_and_phd',
     'Phd': 'Masters_and_phd',
     'Graduate': 'Graduate'
     }

#замэпить этот словарь
df['education_level'].map(d)
```

```python
# функция, применимая не к 1 столбцу, а ко всему датафрейму
#накинем phd-ам тренировочных часов

def new_training_hours(row): # на вход строка датафрейма
  if row['education_level'] == 'PhD':
    return row['training_hours'] + 1000
  return row['training_hours']
```

```python
df['new_training_hours'] = df.apply(new_training_hours, axis=1)
```

# Джойны

```python
# подтянуть данные из стороннего датасета
# например, создадим таблицу с возрастом и подтянем ее
```

```python
age = df[['enrollee_id']].copy()
age['age'] = 30
```

```python
age
```

В пандас join делается через функцию merge

```python
df = df.merge(age, how='left', on='enrollee_id')
#left - сохранить все строки
# innerjoin - пересечение таблиц
# on - по какому ключу делать join
```

```python
# важно проверить не появились ли дубликаты после джойна
len(df)
```

```python
age['enrollee_id'].drop_duplicates() # вот так выбрасываются дубликаты
```

```python
# если в таблице есть задубливания, перед тем, как джойнить, нужно сгруппировать данные
# например, если для задубленных айди были указаны разные возраста, можно взять меньший

age.groupby('enrollee_id')['age'].min()
```
